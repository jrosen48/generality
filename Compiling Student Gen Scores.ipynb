```python
import pandas as pd

the_dir = '/Users/joshuarosenberg/Documents/Gen/'
```

```python
##### this reads in the three files in the_dir, and selects and renames codes
```

```python
df1 = pd.read_csv(the_dir + 'Big Generality Coding 7C 8-3.csv')
df2 = pd.read_csv(the_dir + 'Big Generality Coding 7C 12-2.csv')
df3 = pd.read_csv(the_dir + 'Big Generality Coding 8B 6-2.csv')

df1 = df1[['id', 'code']]
df2 = df2[['id', 'code']]
df3 = df3[['id', 'code']]

df1 = df1.rename(columns = {'code': '7C_8-3'})
df2 = df2.rename(columns = {'code': '7C_12-2'})
df3 = df3.rename(columns = {'code': '8B_6-2'})
```

```python
##### merge all three and different combinations (inner means to only merge if ID is in both)
```

```python
first_two = pd.merge(df1, df2, how = 'inner', on = 'id')
last_two = pd.merge(df2, df3, how = 'inner', on = 'id')
first_and_last = pd.merge(df1, df3, how = 'inner', on = 'id')
all_three = pd.merge(pd.merge(df2, df3, how = 'inner', on = 'id'), df1, how = 'inner', on='id')
```

```python
##### Look at number of observations / students in each
```

```python
print('First two: ' + str(first_two.shape[0]) + ' observations')
print('Last two: ' + str(last_two.shape[0]) + ' observations')
print('First and last: ' + str(first_and_last.shape[0]) + ' observations')
print('All three: ' + str(all_three.shape[0]) + ' observations')
```

```python
##### Save merged files
```

```python
last_two.to_csv(the_dir + 'last_two.csv', index = False)
first_and_last.to_csv(the_dir + 'first_and_last.csv', index = False)
first_two.to_csv(the_dir + 'first_two.csv', index = False)
all_three.to_csv(the_dir + 'all_three.csv', index = False)
```

```python
##### this way looks at unique ids, then merges the three files on the data frame with unique ids using how = 'left'
```

```python
frames = pd.concat([df1['id'], df2['id'], df3['id']])
out_df = pd.DataFrame(frames.unique(), columns = ['id'])
out_df = pd.merge(out_df, df1, how = 'left', on = 'id') 
out_df = pd.merge(out_df, df2, how = 'left', on = 'id') 
out_df = pd.merge(out_df, df3, how = 'left', on = 'id') 
out_df.to_csv(the_dir + "all_observations.csv", index = False)
```

```python
#make a list of selected student IDs and scores
def IDs():
    idlist = []
    with open(the_dir + coder_workbook, 'r') as csvfile: 
        dialect = csv.excel
        reader = csv.reader(csvfile)
        for row in reader: 
            idlist = idlist + [row[0]]
    return idlist

def scores():
    scorelist = []
    with open(the_dir + coder_workbook, 'r') as csvfile:
        dialect = csv.excel
        reader = csv.reader(csvfile)
        for row in reader:
            scorelist = scorelist + [row[3]]     
    return scorelist
```

```python
scores()
```

```python
#flatten the lists 
f_IDs = IDs()
f_scores = scores()
```

#### Run the next cells, and check for any entry errors:

```python
#compiling ids dictionary
ids = {}
for n, v in zip(f_IDs, f_scores):
    ids[n] = v 
```

```python
#Defining error-checking functions
#Auto-checking for blank entry errors
def blank_entry():
    clean = True
    for n, v in ids.items():
        for entry in v:
            if entry is '':
                clean = False
                print (n, 'has a blank entry')
    if clean:
        print ('No blank entry errors!')
```
